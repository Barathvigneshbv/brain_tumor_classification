```python
open_brackets = ["[","{","("]
close_brackets = ["]","}",")"]

def check_string(string):
  stack = []
  for i in string:
    if i in open_brackets:
      stack.append(i)
    elif i in close_brackets:
      pos = close_brackets.index(i)
      if ((len(stack) > 0) and
        (open_brackets[pos] == stack[len(stack)-1])):
        stack.pop()
      else:
        return "false"
  if len(stack) == 0:
    return "true"
  else:
    return "false"
```

```python
string1 = "(A + B * C + D)"
print(check_string(string1))

string2 = "(A + B) * (C + D)"
print(check_string(string2))

string3 = "(A+B]"
print(check_string(string3))
```

```python
class Solution:
    def maxProfit(self, prices):
        n=len(prices)
        dp=dp1=dp2=dp3=mx_sofar=0
        for i in range(n-2,-1,-1):
            mx_sofar=max(mx_sofar,prices[i+1]+dp3)
            dp=max(dp1,mx_sofar-prices[i])
            dp1,dp2,dp3=dp,dp1,dp2
        return dp
```
